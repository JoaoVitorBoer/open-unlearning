```python
import pandas as pd
```

```python
ROUGE_COLUMNS = [
    "forget_knowmem_ROUGE",
    "forget_verbmem_ROUGE",
    "retain_knowmem_ROUGE",
]

def format_rouge_columns(df, precision=2, columns=None, scale=1000):
    """Return a Styler with scaled ROUGE columns formatted to fixed precision."""
    if columns is None:
        columns = ROUGE_COLUMNS
    def _fmt(value):
        return f"{value * scale:.{precision}f}"
    format_map = {
        col: _fmt for col in columns if col in df.columns
    }
    return df.style.format(format_map)
```

```python
df = pd.read_csv('C:\\Users\\Usuario\\Desktop\\Repos\\open-unlearning\\results\\grad_diff.csv')
df = format_rouge_columns(df, precision=2)
```
